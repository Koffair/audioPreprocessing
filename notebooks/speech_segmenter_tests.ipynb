# Preprocessing Audio Files
## Slicing up raw data

```python
import os
from pydub import AudioSegment

data_path = "../data/raw"
mp3s = [os.path.join(data_path, f) for f in os.listdir(data_path) if os.path.isfile(os.path.join(data_path, f))]
```

```python
out_path = "../data/interim/raw_slices"
minute = 60 * 10000
for f in mp3s:
    fname = f.split("/")[-1].split(".")[0]
    programme = AudioSegment.from_mp3(f)
    print(fname, len(programme))
    for i in range(0, len(programme), minute):
        segment = programme[i:i+minute]
        segment.export(os.path.join(out_path, fname + f"_{i}.mp3"), format="mp3")
        print(f"just saved {f} segment {i}")
```

Getting segment info.

```python
import os
from inaSpeechSegmenter import Segmenter
from inaSpeechSegmenter.export_funcs import seg2csv, seg2textgrid
```

```python
data_path = "../data/interim/raw_slices"
mp3s = [os.path.join(data_path, f) for f in os.listdir(data_path) if os.path.isfile(os.path.join(data_path, f))]
seg = Segmenter()
```

```python
import warnings
warnings.filterwarnings('ignore')

speech_path = "../data/interim/speech_segments"

for f in mp3s:
    segmentation = seg(f)
    programme = AudioSegment.from_mp3(f)
    j = 0
    print(f)
    for segment in segmentation:
        if segment[0] == "male" or segment[0] == "female":
            j += 1
            outfile = f.split("/")[-1].split(".")[0]
            outfile = f"{outfile}_{j}.mp3"
            outfile = os.path.join(speech_path, outfile)
            #TODO: introduce overlapping
            onset = segment[1] * 1000
            offset = segment[2] * 1000
            speech = programme[onset:offset]
            speech.export(outfile, format="mp3")
            print(f"saved\t{f}\t{outfile}")
```
